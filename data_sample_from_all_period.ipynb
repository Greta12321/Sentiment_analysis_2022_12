## Taking the sample of tweets from each day from the war beginning

The data is dowloaded from Kaggle (link below) and sample and taken from each day. Only the four days in December (18,19,20,21) are not available. In total, there are tweets from 130 days starting from 2022-08-19 to the 2023-01-01. Firstly, 800 of tweets are taken from each day, when only english language tweets are filtered. In total there are 58,739 tweets for sentiment analysis.

The sample is taken in local notebook, the data is saved as csv file and when uploaded to github.

https://www.kaggle.com/datasets/bwandowando/ukraine-russian-crisis-twitter-dataset-1-2-m-rows?select=UkraineWar

```python
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
```

```python
N = 800
df = pd.concat([pd.read_csv(f).sample(N) for f in csv_files], ignore_index=True)
```

```python
df
```

```python
for col in df.columns:
    print(col)
```

```python
df_columns = df[['username', 'extractedts', 'text', 'language']]
```

```python
df_columns.rename(columns = {'text':'tweet', 'extractedts':'created_at'}, inplace = True)
```

```python
df_columns
```

```python
df_columns.sort_values(by=['created_at'])
```

```python
df_columns["language"].value_counts()
```

```python
#Leaving only english language tweets
df_columns = df_columns.query('language == "en" or language == "uk"') 
```

```python
df_columns["language"].value_counts()
```

```python
df_columns['created_at'] = pd.to_datetime(df_columns['created_at']).dt.date
```

```python
df_columns.sort_values(by=['created_at'])
```

```python
df_columns['created_at'].value_counts()
```

```python
df_columns.to_csv('russia_war_data_sample.csv')
```
